```python
# Style block with Rule
Fragment(
    Style(
        Rule(".demo")
            .background_color("#f0f0f0")
            .padding("1rem")
            .border_radius("8px")
            .color("blue"),
        Rule(".demo h2").color("red"),
    ),
    Div(
        H2("Styled with Rule"),
        P("This uses a Style block with Rule selectors."),
    ).cls("demo"),
)
```

```python
# Inline style with .style()
Div(
    H2("Styled inline"),
    P("This uses the .style() method with InlineStyle."),
).style(
    InlineStyle()
        .background_color("#e8f4e8")
        .padding("1rem")
        .border_radius("8px")
        .color("green")
)
```

```python
# Custom component with _repr_html_
class Card:
    def __init__(self, title, body):
        self.title = title
        self.body = body

    def __ztml_render__(self):
        return Div(H2(self.title), P(self.body)).cls("card")

    def _repr_html_(self):
        return render(self)

Card("Custom Component", "This renders via _repr_html_ too!")
```
